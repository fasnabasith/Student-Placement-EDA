# **BAR PLOT AND HISTOGRAM**

_**Count of Students By PlacementStatus:**_

```python
Placement_counts=df["PlacementStatus"].value_counts()
Placement_counts
```

```python
sns.countplot(x='PlacementStatus', data=df, palette='viridis')
plt.xlabel('Placement Status')
plt.ylabel('Count')
plt.title('Placement Status Count')
plt.show()
```

```python
# Count plot for PlacementTraining
sns.countplot(x='PlacementTraining', data=df, palette='rocket')
plt.xlabel('Placement Training')
plt.ylabel('Count')
plt.title('Placement Training Count')
plt.show()
```

```python
# Count plot for ExtracurricularActivities
sns.countplot(x='ExtracurricularActivities', data=df,palette='icefire' )
plt.title('Extracurricular Activities Count')
plt.show()
```

```python
custom_palette = {"Placed": "darkblue", "NotPlaced": "lightblue"}
plt.figure(figsize=(12, 6))
sns.countplot(x='CGPA',hue='PlacementStatus', data=df, palette=custom_palette)
plt.title('CGPA vs Placement Status"', fontsize=16)
plt.xlabel('CGPA Range', fontsize=12)
plt.ylabel('Count', fontsize=8)
plt.show()
```

```python
pd.crosstab(df["PlacementTraining"], df["PlacementStatus"]).plot(kind="bar", stacked=True, color=["skyblue", "darkblue"])
plt.title("Effect of Placement Training on Placement")
plt.xlabel("Placement Training")
plt.ylabel("Count")
plt.legend(["Not Placed", "Placed"])
plt.show()
```

Students who participated in placement training had a higher chance of getting placed compared to those who didn’t. This suggests that placement training plays a significant role in improving students' job opportunities

```python
# Histogram for CGPA
plt.figure(figsize=(8, 5))
sns.histplot(df['CGPA'], bins=20, kde=True, color='dodgerblue')
plt.title('Distribution of CGPA', fontsize=16, weight='bold')
plt.xlabel('CGPA', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# AptitudeTestScore
plt.hist(df['AptitudeTestScore'], bins=20, color='dodgerblue')
plt.title('Distribution of Aptitude Test Score')
plt.xlabel('Aptitude Test Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# SoftSkillsRating
plt.hist(df['SoftSkillsRating'])
plt.title('Distribution of Soft Skills Rating')
plt.xlabel('Soft Skills Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='CGPA', hue='PlacementStatus', multiple='stack', palette='Set2', bins=20)
plt.title('Histogram of CGPA by Placement Status')
plt.xlabel('CGPA')
plt.ylabel('Count')
plt.show()
```

```python
# SoftSkillsRating histogram by PlacementStatus
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='SoftSkillsRating', hue='PlacementStatus', multiple='stack', palette='Set3', bins=20)
plt.title('Histogram of Soft Skills Rating by Placement Status')
plt.xlabel('Soft Skills Rating')
plt.ylabel('Count')
plt.show()


```

```python
# AptitudeTestScore histogram by PlacementStatus
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='AptitudeTestScore', hue='PlacementStatus', multiple='stack', palette='Set1', bins=20)
plt.title('Histogram of Aptitude Test Score by Placement Status')
plt.xlabel('Aptitude Test Score')
plt.ylabel('Count')
plt.show()
```
